```
# Janome動作サンプル
from janome.tokenizer import Tokenizer
j_t = Tokenizer()
text = "機械学習が好きです。"
for token in j_t.tokenize(text):
    print(token)
```

```
# 単語分割の関数
def tokenizer_janome(text):
    return [token for token in j_t.tokenize(text, wakati=True)]

text = "機械学習が好きです。"
print(tokenizer_janome(text))
```
